### Getting started

```python
# import useful packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
#import cufflinks as cf

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# define a pipeline for polynomial regression
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         LinearRegression(**kwargs))
```

Next, let us define a function to make some toy data and display the scatter plot:

```python
# make some example data
def make_data(N=30, err=0.8, rseed=1):
    # randomly sample the data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2
    y = 10 - 1. / (X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X, y
```

```python
X, y = make_data()
plt.scatter(X.ravel(), y, s=40)
plt.grid(True)
```

```python
df = pd.DataFrame({'X':X.ravel(),'Y':y})
df.head()
```

As our first modeling attempt, let us try the linear regression, i.e., a PolynomialRegression of degree 1. Note the usual sequence of operations we take - first we call the **fit()** function to train the model with the data, and then do a **predict()** for new data - in this case, it is the entire range of $X\in[-1.0, 1.0]$. The $R^2$ is then reported for the training data.

```python
# Fit the model to data
# GENERATE the m and the b in the equation
model1 = PolynomialRegression(1)
model1.fit(X, y)

#Create a bunch of x values and plot them against predicted y
#ie: plt.plot(x_examples, y_from_x)
xfit = np.linspace(-0.1, 1.0, 1000)[:, None]
plt.scatter(X.ravel(), y, s=40)
plt.plot(xfit.ravel(), model1.predict(xfit), color='gray')

plt.axis([-0.1, 1.0, -2, 14]); plt.grid(True)
plt.title('Fitting result', size=14);
print 'R2=', model1.score(X, y)
```

Let's increase the complexity of our model and try a quadradic one (degree of 2). 

```python
#Generate model with m1 and m2 and m3
model2 = PolynomialRegression(2)
model2.fit(X, y)
#Plot predicted
plt.scatter(X.ravel(), y, s=40)
plt.plot(xfit.ravel(), model2.predict(xfit), color='gray')
#Tinker and test 
plt.axis([-0.1, 1.0, -2, 14])
plt.title('Fitting result', size=14);
print 'R2=', model2.score(X, y)
```

The following code generates two subfigures side by side, and print out the model information:

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
ax[0].scatter(X.ravel(), y, s=40)
ax[0].plot(xfit.ravel(), model1.predict(xfit), color='gray')
ax[0].axis([-0.1, 1.0, -2, 14])
ax[0].set_title('order=1, $R^2$={}'.format(model1.score(X,y)), size=14)
ax[1].scatter(X.ravel(), y, s=40)
ax[1].plot(xfit.ravel(), model2.predict(xfit), color='gray')
ax[1].axis([-0.1, 1.0, -2, 14])
ax[1].set_title('order=2, $R^2$={}'.format(model2.score(X,y)), size=14);
```

```python
model3 = PolynomialRegression(3)
model3.fit(X, y)

plt.scatter(X.ravel(), y, s=40)
plt.plot(xfit.ravel(), model3.predict(xfit), color='grey')

plt.axis([-0.1, 1.0, -2, 14])
plt.title('Fitting result', size=14);
print 'R2=', model3.score(X, y)

#imporovement in R from degree 2!
```

Follow the example given above. For polynomial degrees from 1 to 8, generate the fitting models and visualize them. Report the $R^2$ scores. Which models give the best $R^2$ scores?

```python
r2s = [] #lets see how R2s change with degree increase as well
task1Models = []  #and save our trained models for later
f, ax = plt.subplots(4, 2, figsize=(15,20))

def plotModel(degree, a):
    model = PolynomialRegression(degree).fit(X, y)
    ax.ravel()[a].scatter(X.ravel(), y, s=40)
    ax.ravel()[a].plot(xfit.ravel(), model.predict(xfit), color='grey')
    
    task1Models.append(model)
    r2s.append(model.score(X, y))
    title = "Degree = {0}       R2 = {1}".format(degree, model.score(X, y))
    ax.ravel()[a].set_title(title)
    
for d in range(8):
    plotModel(d+1, d)
```

### Task 2
Now, generate 20 new data instances for testing. 

```python
X2, y2 = make_data(20, rseed=42)
```

**Task 2a.** Reuse the trained models in Task 1. Regenerate the plots that display the *testing* data, the fitting curves, and the fitting scores on the *testing* data. 

```python
#Lets use the models we stored in 'task1Models' to see how they perform against testing data
f, ax = plt.subplots(4, 2, figsize=(15,18))

#Lets write a function that displays the testing data against the trained curve for each model
def plotTestData(models, degree, a):
    model = models[a]
    train = ax.ravel()[a].scatter(X.ravel(), y, s=40, color='blue')  #training points
    test = ax.ravel()[a].scatter(X2.ravel(), y2, s=40, color='red')  #testing points
    ax.ravel()[a].plot(xfit.ravel(), model.predict(xfit), color='grey')         
    
    #Lets also make it easier to differentiate between train and test scores and datapoints
    ax.ravel()[a].set_title('Degree {2}  |  Train R2={0:.3f}  Test R2={1:.3f}'.format(model.score(X, y), model.score(X2, y2), degree))
    ax.ravel()[a].legend([train, test], ['Train', 'Test'], loc=4)
    
for d in range(8):
    plotTestData(task1Models, d+1, d)
```

**Task 2b.** For the model degree ranged from 1 to 11 (inclusive), retrain (using X and y) and test a model (using X2 and y2). Use two lists to record the training and testing scores of each model. Plot out the lists across the degree range. 

```python
trainScores = []
testScores = []

#Lets make a method that retrains models on degree and populates our R2 lists
def trainVtest(degree):
    model = PolynomialRegression(degree).fit(X, y)
    trainScores.append(model.score(X, y))
    testScores.append(model.score(X2, y2))

for d in range(1, 12):
    trainVtest(d)

#Lets now plot the results
train = plt.scatter(range(1,12), trainScores, color='blue')
test = plt.scatter(range(1,12), testScores, color='red')
plt.xlabel('Degree')
plt.ylabel('R2')
plt.title("Training and Testing against Degree")
plt.legend([train, test], ['Train', 'Test'], loc='best')
plt.show()
```
